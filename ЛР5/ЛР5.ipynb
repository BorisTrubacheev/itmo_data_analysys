# Задача 1

Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
indices = list(range(df_length))
q66 = np.quantile(indices, 0.66)

np.random.shuffle(indices)
print('Перемешанные индексы: ', indices)
```

```python
train = df[indices <= q66]
test = df[indices > q66]

print('Размер тренировочной выборки: ', len(train))
print('Размер тестовой выборки: ', len(test))
```

# Задача 2
Разделить таблицу данных на четыре одинаковых части.

```python
indices = list(range(df_length))
q25 = np.quantile(indices, 0.25)
q50 = np.quantile(indices, 0.5)
q75 = np.quantile(indices, 0.75)

np.random.shuffle(indices)
print('Перемешанные индексы: ', indices)
```

```python
part1 = df[indices <= q25]
part2 = df[(indices > q25) & (indices <= q50)]
part3 = df[(indices > q50) & (indices <= q75)]
part4 = df[indices > q75]

print('Размер 1 выборки: ', len(part1))
print('Размер 2 выборки: ', len(part2))
print('Размер 3 выборки: ', len(part3))
print('Размер 4 выборки: ', len(part4))
```

# Задача 3
Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
from math import ceil
proportion = 0.7

df['strata'] = df['COKE'].astype(str) + '_' + df['TAB'].astype(str)

indices = np.array(range(df_length))
np.random.shuffle(indices)

train_indices = []


for strata_name, group in df.groupby('strata'):
    group_indices = group.index.to_list()
    n_train = int((len(group)) * proportion)
    np.random.shuffle(group_indices)
    train_indices.extend(group_indices[:n_train])

train = df.loc[train_indices]
test = df.drop(train_indices)
```

```python
print("Исходные:\n", df[['COKE', 'TAB']].value_counts())
print("\nTrain:\n", train[['COKE', 'TAB']].value_counts())
print("\nTest:\n", test[['COKE', 'TAB']].value_counts())
```
